```python
import pandas as pd
df=pd.read_csv('/content/Book1.csv')
df
```

```python
target='pass or fail'
print(df[target])
```

```python
x=df.drop(target,axis=1)
print(x)
```

```python
w0=1
w1=3
w2=1
w3=1
w4=1
w5=1
z=w0*1+w1*x["study hours per week"]+w2*x["class attendance %"]+w3*x["previous exam score"]+w4*x["participation in group study"]+w5*x["subission of assignments"]
print(z)
yp=[]
for i in range(10):
  if z[i]>0:
    yp.append(1)
    #print(yp)
  else:
    yp.append(0)
   # print(yp)
```

```python
print(yp)
```

```python
error=df[target]-yp
print(error)
```

```python
mean=error.mean()
print(mean)
```

```python
e_update=w0+0.01*mean
print(e_update)
e_update1=w1+0.01*mean
print(e_update1)
e_update2=w2+0.01*mean
print(e_update2)
e_update3=w3+0.01*mean
print(e_update3)
e_update4=w4+0.01*mean
print(e_update4)
e_update5=w5+0.01*mean
print(e_update5)
```
